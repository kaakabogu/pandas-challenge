### Distirct Summary
Perform the necessary calculations and then create a high-level snapshot of the district's key metrics in a DataFrame.

Include the following:

- Total number of unique schools

- Total students

- Total budget

- Average math score

- Average reading score

- % passing math (the percentage of students who passed math)

- % passing reading (the percentage of students who passed reading)

- % overall passing (the percentage of students who passed math AND reading)

```python
#combine th 2 dataset into one dataset
district_summary = pd.merge(students_data_df, schools_data_df)
district_summary.head()
```

```python
#calculate the total number of unqiue schools
school_counts = district_summary['school_name'].nunique()
school_counts
```

```python
# Total students
student_counts = district_summary['student_name'].count()
student_counts
```

```python
# Total budget
total_budget = schools_data_df['budget'].sum()
total_budget
```

```python
# calculate the average math score
avg_math_score = district_summary['math_score'].mean()
avg_math_score
```

```python
# calculate the average reading score
avg_reading_score = district_summary['reading_score'].mean()
avg_reading_score
```

```python
passing_math_count = district_summary[(district_summary['math_score'] >= 70)].count()['student_name']
passing_math_percentage = passing_math_count / float(student_counts) * 100
passing_math_percentage
```

```python
passing_read_count = district_summary[(district_summary['reading_score'] >= 70)].count()['student_name']
passing_read_percentage = passing_read_count / float(student_counts) * 100
passing_read_percentage
```

```python
overall_passing_count = district_summary[(district_summary['math_score']>=70) & (district_summary['reading_score']>=70)].count()['student_name']
overall_passing_rate = overall_passing_count/float(student_counts)*100
overall_passing_rate
```

```python
summary_district = pd.DataFrame({
    'Total schools': school_counts, 
    'Total students':f"{student_counts:,}", 
    'Total budget': f"${total_budget:,.2f}",
    'Average math score': f"{avg_math_score:.6f}",
    'Average read score': f"{avg_reading_score:.5f}",
    '% passing math': f"{passing_math_percentage:.6f}",
    '% passing reading': f"{passing_read_percentage:.6f}",
    '% overall passing': f"{overall_passing_rate:.6f}"}, 
    index=[0])
summary_district
```

```python
# total_budget
total_school_budget = district_summary.groupby(["school_name"]).mean()["budget"]
total_school_budget
```

```python
# % passing math (the percentage of students who passed math)
passing_math_count = district_summary[(district_summary['math_score'] >= 70)].groupby('school_name')['Student ID'].count()
passing_math_percentage = passing_math_count / (school_students_count) * 100
passing_math_percentage
```

```python
# % passing reading (the percentage of students who passed reading)
passing_reading_count = district_summary[(district_summary['reading_score'] >= 70)].groupby('school_name')['Student ID'].count()
passing_reading_percentage = passing_reading_count / (school_students_count) * 100
passing_reading_percentage
```

```python
# % overall passing (the percentage of students who passed math AND reading)
overall_passing_count = district_summary[(district_summary['math_score'] >= 70) & (district_summary['reading_score']>69)].groupby('school_name')['Student ID'].count()
overall_passing_percentage = overall_passing_count / (school_students_count) * 100
overall_passing_percentage
```

```python
per_school_summary= pd.DataFrame({'School Type': school_type,
                                  'Total students': school_students_count,
                                  'Total school budget': total_school_budget,
                                  'Per student budget': per_student_budget, 
                                  'Average math score': avg_math_score,
                                  'Average reading score': avg_reading_score,
                                  '% passing math': passing_math_percentage,
                                  '% passing reading': passing_reading_percentage,
                                  '% overall passing': overall_passing_percentage})

per_school_summary = per_school_summary[['School Type','Total students',
                                        'Total school budget','Per student budget',
                                        'Average math score','Average reading score',
                                        '% passing math','% passing reading', '% overall passing']]
per_school_summary
```

### Scores by School Spending 
Use pd.cut with the provided code to bin the data by the spending ranges 
Use the code provided to calculate the averages 
Create the spending_summary DataFrame using the binned and averaged spending data 

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Per student budget'], bins = spending_bins, labels = labels)

#group by spending
by_school_spending = school_spending_df.groupby('Spending Ranges (Per Student)')

spending_math_scores = by_school_spending.mean()["Average math score"]
spending_reading_scores = by_school_spending.mean()["Average reading score"]
spending_passing_math = by_school_spending.mean()["% passing math"]
spending_passing_reading = by_school_spending.mean()["% passing reading"]
overall_passing_spending = by_school_spending.mean()["% overall passing"]


# df build            
scores_by_spend = pd.DataFrame({
    "Average math score": spending_math_scores,
    "Average reading score": spending_reading_scores,
    "% passing math": spending_passing_math,
    "% passing reading": spending_passing_reading,
    "% overall passing": overall_passing_spending
            
})
            
#reorder columns
scores_by_spend = scores_by_spend[[
    "Average math score",
    "Average reading score",
    "% passing math",
    "% passing reading",
    "% overall passing"
]]

scores_by_spend.index.name = "Spending Ranges (Per Student)"

scores_by_spend







```

### Scores by School Size 
Use pd.cut with the provided code to bin the data by the school sizes 
Use the code provided to calculate the averages 
Create the size_summary DataFrame using the binned and averaged size data 

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
# referencing the school_spending_df dateframe defined from score by spending calculations

per_school_summary["School Size"] =  pd.cut(per_school_summary['Total students'], bins = size_bins, labels = labels)

#group by spending
by_school_size = per_school_summary.groupby('School Size')

size_math_scores = by_school_size.mean()["Average math score"]
size_reading_scores = by_school_size.mean()["Average reading score"]
size_passing_math = by_school_size.mean()["% passing math"]
size_passing_reading = by_school_size.mean()["% passing reading"]
size_overall_passing = by_school_size.mean()["% overall passing"]


# df build            
scores_by_size = pd.DataFrame({
    "Average math score": size_math_scores,
    "Average reading score": size_reading_scores,
    "% passing math": size_passing_math,
    "% passing reading": size_passing_reading,
    "% overall passing": size_overall_passing
            
})
            
#reorder columns
scores_by_size = scores_by_size[[
    "Average math score",
    "Average reading score",
    "% passing math",
    "% passing reading",
    "% overall passing"
]]

scores_by_size.index.name = "School Size"

scores_by_size



```
